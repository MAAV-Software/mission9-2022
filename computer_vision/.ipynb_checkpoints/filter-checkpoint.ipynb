```python
def intersection(m1: float, b1: float, m2: float, b2: float):
    # Consider y to be equal and solve for x
    # Solve:
    #   m1 * x + b1 = m2 * x + b2
    x = (b2 - b1) / (m1 - m2)
    # Use the value of x to calculate y
    y = m1 * x + b1

    return int(round(x)), int(round(y))
```

```python
def hough_inter(theta1, rho1, theta2, rho2):
    A = np.array([[cos(theta1), sin(theta1)], 
                  [cos(theta2), sin(theta2)]])
    b = np.array([rho1, rho2])
    return np.linalg.lstsq(A, b)[0] # use lstsq to solve Ax = b, not inv() which is unstable
```

```python
def order_points(pts):
    rect = np.zeros((4, 2), dtype="float32")

    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]
    rect[2] = pts[np.argmax(s)]

    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]
    rect[3] = pts[np.argmax(diff)]

    return rect
```

```python
#Given theta in radians
def close_to_verticle(theta):
    #verticle lines are integer multiples of pi/2
    PI_2 = math.pi / 2
    
    num = (abs(theta) / PI_2)
    closest_pi_over_2 = round(num)
    
    threshold = 0.1
    if abs(num - closest_pi_over_2) < threshold:
        return True
    
#Given theta in radians
def close_to_horizontal(theta):
    #horizontal lines are integer multiples of pi
    PI = math.pi
    
    num = (abs(theta) / PI)
    closest_pi = round(num)
    
    threshold = 0.1
    if abs(num - closest_pi) < threshold:
        return True
```

```python
#Finds the lines that bound the mast
#If image parameter is supplied, it will draw the lines on that image
def find_bounding_lines(edges, image = None):
    myMinVertLine = []
    myMaxVertLine = []
    myMinHorizLine = None #r, theta
    myMaxHorizLine = None
    
    lines = cv2.HoughLines(edges, 1, np.pi/180, 60)
    for r_theta in lines:
        arr = np.array(r_theta[0], dtype=np.float64)
        r, theta = arr

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a*r
        y0 = b*r
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        
        #check if lines are sortve vertical or sortve horizontal and then see if they beat the extremes
        if close_to_horizontal(theta):
            if not myMinVertLine:
                print(r)
                myMinVertLine = [(x1, y1), (x2, y2), (r, theta)]
            elif x2+x1 < myMinVertLine[1][0]+myMinVertLine[0][0]:
                print(r)
                myMinVertLine = [(x1, y1), (x2, y2), (r, theta)]
#             if not myMaxVertLine:
#                 print(r)
#                 myMaxVertLine = [(x1, y1), (x2, y2), (r, theta)]
            #elif x2+x1 > myMaxVertLine[1][0]+myMaxVertLine[0][0]:
                print(r)
                #myMaxVertLine = [(x1, y1), (x2, y2), (r, theta)]
        elif close_to_verticle(theta):
            if not myMinHorizLine:
                myMinHorizLine = (r, theta)
            elif abs(r) < abs(myMinHorizLine[0]):
                myMinHorizLine = (r, theta)
                
            if not myMaxHorizLine:
                myMaxHorizLine = (r, theta)
            elif abs(r) > abs(myMaxHorizLine[0]):
                myMaxHorizLine = (r, theta)
    
    if image is not None:
        cv2.line(image, myMinVertLine[0], myMinVertLine[1], (0, 0, 255), 10)
        cv2.line(image, myMaxVertLine[0], myMaxVertLine[1], (0, 0, 255), 10)
        cv2.line(image, myMinHorizLine[0], myMinHorizLine[1], (0, 0, 255), 10)
        cv2.line(image, myMaxHorizLine[0], myMaxHorizLine[1], (0, 0, 255), 10)
    
    return myMinVertLine[2], (0, 0), myMinHorizLine, myMaxHorizLine
```
