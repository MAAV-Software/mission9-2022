```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
image = cv.imread("/Users/dermotmolony/Desktop/Screen Shot 2022-10-09 at 2.29.54 PM.png")
#src = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
# Convert BGR to HSV
hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)
# define range of blue color in HSV
lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])
# Threshold the HSV image to get only blue colors
mask = cv.inRange(hsv, lower_blue, upper_blue)
# Bitwise-AND mask and original image
res = cv.bitwise_and(image,image, mask= mask)
# cv.imshow('frame',frame)
# cv.imshow('mask',mask)
# cv.imshow('res',res)


src = cv.cvtColor(res, cv.COLOR_BGR2RGB)
plt.imshow(src)
plt.show()
grayscale = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(grayscale)
plt.show()
grayscale = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(grayscale)
plt.show()
```
